```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

X = np.array([-7.0, -4.0, -1.0, 2.0, 5.0], dtype=np.float32)
Y = np.array([3.0, 6.0, 9.0, 12.0, 15.0], dtype=np.float32)
plt.scatter(X, Y)  # plot the data points

# Reshape X and Y to (num_samples, num_features)
X = tf.constant(X)
Y = tf.constant(Y)
X = tf.reshape(X, (-1, 1))  # Reshape to (5, 1)
Y = tf.reshape(Y, (-1, 1))  # Reshape to (5, 1)

tf.random.set_seed(42)
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])
model.compile(loss=tf.keras.losses.mae, optimizer=tf.keras.optimizers.SGD(), metrics=["mae"])
model.fit(X, Y, epochs=5)

# Predict on new data with the correct shape and type
y_pred = model.predict([[17.0]])
print(y_pred)
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
x_train = np.linspace(0, 10, 100)  # 100 training samples
y_train = 2 * x_train + 1 + np.random.normal(0, 1, 100)  # Linear data with noise

x_test = np.linspace(10, 15, 20)  # 20 test samples
y_test = 2 * x_test + 1 + np.random.normal(0, 1, 20)

# Define the plot_predictions function
def plot_predictions(train_data=x_train,
                     train_labels=y_train,
                     test_data=x_test,
                     test_labels=y_test,
                     predictions=None):
    """
    Plots training data and compares predictions to ground truth.
    """
    plt.figure(figsize=(10, 7))
    # Plot training data in blue
    plt.scatter(train_data, train_labels, c="b", label="Training data")
    # Plot test data in green
    plt.scatter(test_data, test_labels, c="g", label="Test data")
    # Plot predictions in red (only if provided)
    if predictions is not None:
        plt.scatter(test_data, predictions, c="r", label="Predictions")
    plt.legend()
    plt.show()  # Display the plot

# Reshaping the data to ensure correct shape
x_train = tf.reshape(x_train, (-1, 1))  # (num_samples, 1 feature)
x_test = tf.reshape(x_test, (-1, 1))    # (num_samples, 1 feature)
y_train = tf.reshape(y_train, (-1, 1))  # (num_samples, 1 target)
y_test = tf.reshape(y_test, (-1, 1))    # (num_samples, 1 target)

# Build and compile model_1 (1 layer)
tf.random.set_seed(42)
model_1 = tf.keras.Sequential([
    tf.keras.layers.Dense(1)  # 1 layer, simple linear regression model
])

model_1.compile(loss=tf.keras.losses.MeanAbsoluteError(),
                optimizer=tf.keras.optimizers.SGD(),
                metrics=["mae"])

# Train the model
model_1.fit(x_train, y_train, epochs=100)

# Predict using the model
y_preds_1 = model_1.predict(x_test)
y_preds_1 = tf.squeeze(y_preds_1)  # Remove single dimensions from the predictions

# Plot the predictions
plot_predictions(train_data=x_train, train_labels=y_train,
                 test_data=x_test, test_labels=y_test, predictions=y_preds_1)

# Calculate and print MAE and MSE for model_1
mae_1 = tf.keras.losses.MeanAbsoluteError()(y_test, y_preds_1)
mse_1 = tf.keras.losses.MeanSquaredError()(y_test, y_preds_1)

print("MAE for model 1:", mae_1.numpy())
print("MSE for model 1:", mse_1.numpy())

# Now you can proceed with model_2

# Build and compile model_2 (2 layers)
model_2 = tf.keras.Sequential([
    tf.keras.layers.Dense(10),  # First hidden layer
    tf.keras.layers.Dense(1)    # Output layer
])

model_2.compile(loss=tf.keras.losses.MeanAbsoluteError(),
                optimizer=tf.keras.optimizers.SGD(),
                metrics=["mae"])

# Train the model
model_2.fit(x_train, y_train, epochs=100)

# Predict using model_2
y_pred_2 = model_2.predict(x_test)
y_pred_2 = tf.squeeze(y_pred_2)  # Remove single dimensions from the predictions

# Calculate MAE and MSE for model_2
mae_2 = tf.keras.losses.MeanAbsoluteError()(y_test, y_pred_2)
mse_2 = tf.keras.losses.MeanSquaredError()(y_test, y_pred_2)

print("MAE for model 2:", mae_2.numpy())
print("MSE for model 2:", mse_2.numpy())
# make and plot redictions of model_2
y_preds=model_2.predict(x_test)
plot_predictions(predictions=y_preds)
# build model 3
# model 3 train for 3 hidden layes and h100 epochs
# 2 layers trained for 500 epochs
model_3=tf.keras.Sequential([
    tf.keras.layers.Dense(100,activation="relu"),# inpu layer

    tf.keras.layers.Dense(1)# output layer
])
model_3.compile(loss=tf.keras.losses.MeanAbsoluteError(),
                optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
                metrics=["mae"])
model_3.fit(x_train,y_train,epochs=500)
predict=model_3.predict(x_test)
plot_predictions(predictions=predict)
mae_3=mae(y_test,y_preds)
mse_3=mse(y_test,y_preds)
mae_3,mse_3





```
